# V_eff

```python
from astropy.cosmology import FlatLambdaCDM
import numpy as np

cosmo = FlatLambdaCDM(H0=70, Om0=0.3)

# Correct field area
field_area_deg2 = 3.11968
field_area_sr = field_area_deg2 * (np.pi / 180)**2

z_min = 6.8
z_max = 7.1

V_eff = cosmo.comoving_volume(z_max).value - cosmo.comoving_volume(z_min).value
V_eff *= field_area_sr / (4 * np.pi)

print(f"V_eff = {V_eff:.2e} Mpc^3")
```
