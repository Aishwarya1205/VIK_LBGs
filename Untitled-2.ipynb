```python
import numpy as np
from astropy.coordinates import SkyCoord
import astropy.units as u

# ── File paths ─────────────────────────────────────────
file_txt = "/Users/aishwarya/Desktop/Trystan_ambiguous_output.txt"
file_cat = "/Users/aishwarya/Desktop/CSV/Matched_candidates_withmags.cat"
output_file = "/Users/aishwarya/Desktop/contact_sources.txt"

# ── Load TXT file (RA, DEC only) ───────────────────────
# Uses usecols=(0,1) to grab only RA and DEC, ignoring the empty trailing columns
# encoding handles any stray whitespace around values
txt_data = np.genfromtxt(
    file_txt,
    delimiter=';',
    skip_header=1,
    usecols=(0, 1),          # ← only read RA and DEC columns
    autostrip=True           # ← strips leading/trailing spaces from each field
)

# Ensure 2D even if single row
txt_data = np.atleast_2d(txt_data)

# Remove any bad rows (NaNs)
txt_data = txt_data[~np.isnan(txt_data).any(axis=1)]

# ── Load CAT file (RA, DEC + mags) ─────────────────────
data_cat = np.genfromtxt(
    file_cat,
    comments='#',
    autostrip=True           # ← same safety measure
)

# Ensure 2D even if single row
data_cat = np.atleast_2d(data_cat)

# Remove bad rows
data_cat = data_cat[~np.isnan(data_cat).any(axis=1)]

# ── Guard: check both catalogs have data ───────────────
if txt_data.shape[0] == 0:
    raise ValueError("TXT file is empty or all rows contained NaNs. Check delimiter and file content.")

if data_cat.shape[0] == 0:
    raise ValueError(
        "CAT file is empty or all rows contained NaNs after filtering.\n"
        "Check: (1) file path is correct, (2) columns are whitespace-separated, "
        "(3) no unexpected characters."
    )

ra_txt  = txt_data[:, 0]
dec_txt = txt_data[:, 1]

ra_cat  = data_cat[:, 0]
dec_cat = data_cat[:, 1]

print(f"TXT sources loaded:  {len(ra_txt)}")
print(f"CAT sources loaded:  {len(ra_cat)}")

# ── Quick sanity check on coordinate ranges ────────────
print(f"RA  range (TXT): {ra_txt.min():.4f} – {ra_txt.max():.4f}")
print(f"DEC range (TXT): {dec_txt.min():.4f} – {dec_txt.max():.4f}")

# ── Create SkyCoord objects ────────────────────────────
coords_txt = SkyCoord(ra=ra_txt * u.deg, dec=dec_txt * u.deg)
coords_cat = SkyCoord(ra=ra_cat * u.deg, dec=dec_cat * u.deg)

# ── Match catalogs ─────────────────────────────────────
idx, d2d, _ = coords_txt.match_to_catalog_sky(coords_cat)

# Matching threshold
max_sep = 1.0 * u.arcsec
match_mask = d2d < max_sep

# ── Extract matched sources ────────────────────────────
matched_cat = data_cat[idx[match_mask]]

print(f"Number of matches:   {len(matched_cat)}")

# ── Save output ────────────────────────────────────────
if len(matched_cat) > 0:
    np.savetxt(output_file, matched_cat, fmt="%.6f")
    print(f"Saved matched sources to: {output_file}")
else:
    print("No matches found within 1 arcsec. Output file not written.")
```

```python
import numpy as np
import matplotlib.pyplot as plt

from astropy.coordinates import SkyCoord
import astropy.units as u
from astropy.cosmology import FlatLambdaCDM
from astropy.io import fits
from astropy.wcs import WCS
from astropy.visualization import ZScaleInterval
from regions import Regions

# ----------------------------
# FILES
# ----------------------------
IMG_FILE = "/Users/aishwarya/Desktop/Lyman_alpha_2/Y_band/Mosaic/mosaic_Y.fits"
WT_FILE  = "/Users/aishwarya/Desktop/Lyman_alpha_2/Y_band/Mosaic/mosaic_Y_weight.fits"
REG_FILE = "/Users/aishwarya/Desktop/DECAM.reg"
CAT_FILE = "/Users/aishwarya/Desktop/contact_sources.txt"

# ----------------------------
# COSMOLOGY
# ----------------------------
COSMO = FlatLambdaCDM(H0=70, Om0=0.3)
z = 6.9
arcsec_per_kpc = COSMO.arcsec_per_kpc_proper(z)

def mpc_to_deg(mpc):
    return (mpc * 1000 * u.kpc * arcsec_per_kpc).to(u.deg).value

# ----------------------------
# QUASAR
# ----------------------------
quasar = SkyCoord(ra=357.138875*u.deg,
                  dec=-30.9028417*u.deg)

# ----------------------------
# LOAD IMAGE
# ----------------------------
img_hdul = fits.open(IMG_FILE, memmap=True)
data = img_hdul[0].data.astype(np.float32)
wcs = WCS(img_hdul[0].header)

# ----------------------------
# LOAD WEIGHT → FOOTPRINT MASK
# ----------------------------
wt = fits.open(WT_FILE)[0].data.astype(np.float32)

threshold = 0.003
mask = (wt > threshold) & np.isfinite(wt)

data_plot = data.copy()
data_plot[~mask] = np.nan   # outside footprint = white

# ----------------------------
# LOAD REGION (DECam footprint)
# ----------------------------
regions = Regions.read(REG_FILE, format='ds9')


# ----------------------------
# LOAD SOURCES + REMOVE QSO
# ----------------------------
cat = np.loadtxt(CAT_FILE)

sources_all = SkyCoord(ra=cat[:,0]*u.deg,
                       dec=cat[:,1]*u.deg)

# Compute separation
sep = sources_all.separation(quasar)

# Remove anything within 3 arcsec of QSO
mask_not_qso = sep > 1 * u.arcsec

sources = sources_all[mask_not_qso]

# Convert to pixel AFTER filtering
x_src, y_src = wcs.world_to_pixel(sources)
x_qso, y_qso = wcs.world_to_pixel(quasar)

# Debug check
print("Closest source distance (arcsec):", np.min(sep.arcsec))
print("Removed:", np.sum(~mask_not_qso))
# ----------------------------
# RADII
# ----------------------------
r5 = mpc_to_deg(5)
r10 = mpc_to_deg(10)

# convert to pixels
pixel_scale = np.abs(wcs.pixel_scale_matrix[1,1])  # deg/pixel
r5_pix = r5 / pixel_scale
r10_pix = r10 / pixel_scale

# ----------------------------
# IMAGE SCALING (paper style)
# ----------------------------
zscale = ZScaleInterval()
vmin, vmax = zscale.get_limits(data_plot)

# ----------------------------
# PLOT
# ----------------------------
fig = plt.figure(figsize=(6,6))
ax = plt.subplot(projection=wcs)

ax.set_facecolor('white')

# image
ax.imshow(data_plot,
          origin='lower',
          cmap='gray_r',
          vmin=vmin,
          vmax=vmax)

# ----------------------------
# FOOTPRINT (exact DS9 region)
# ----------------------------
for reg in regions:
    pix_reg = reg.to_pixel(wcs)
    pix_reg.plot(ax=ax, color='black', lw=2)

# ----------------------------
# RINGS
# ----------------------------
circle_outer = plt.Circle((x_qso, y_qso),
                          r10_pix,
                          color='black',
                          fill=False,
                          linewidth=2)

circle_inner = plt.Circle((x_qso, y_qso),
                          r5_pix,
                          color='black',
                          fill=False,
                          linestyle=':',
                          linewidth=2)

# central shaded region
circle_core = plt.Circle((x_qso, y_qso),
                         r5_pix*0.35,
                         color='gray',
                         alpha=0.3)

ax.add_patch(circle_core)
ax.add_patch(circle_outer)
ax.add_patch(circle_inner)

# ----------------------------
# SOURCES
# ----------------------------
ax.scatter(x_src, y_src,
           s=25,
           color='red',
           edgecolors='none',
           zorder=5)

# quasar
ax.scatter(x_qso, y_qso,
           marker='*',
           s=100,
           color='black',
           zorder=6)

# ----------------------------
# AXES STYLE (paper-like)
# ----------------------------
ax.set_xlabel("RA")
ax.set_ylabel("DEC")

ax.coords[0].set_major_formatter('hh:mm')
ax.coords[1].set_major_formatter('dd:mm')

ax.minorticks_on()
ax.tick_params(which='both', direction='in', width=1.2)
ax.tick_params(which='major', length=4)

ax.grid(False)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

from astropy.coordinates import SkyCoord
import astropy.units as u
from astropy.cosmology import FlatLambdaCDM
from astropy.io import fits
from astropy.wcs import WCS
from astropy.visualization import ZScaleInterval
from regions import Regions

# ----------------------------
# FILES
# ----------------------------
IMG_FILE = "/Users/aishwarya/Desktop/Lyman_alpha_2/Y_band/Mosaic/mosaic_Y.fits"
WT_FILE  = "/Users/aishwarya/Desktop/Lyman_alpha_2/Y_band/Mosaic/mosaic_Y_weight.fits"
REG_FILE = "/Users/aishwarya/Desktop/DECAM.reg"
CAT_FILE = "/Users/aishwarya/Desktop/contact_sources.txt"

# ----------------------------
# COSMOLOGY
# ----------------------------
COSMO = FlatLambdaCDM(H0=70, Om0=0.3)
z = 6.9
arcsec_per_kpc = COSMO.arcsec_per_kpc_proper(z)

def mpc_to_deg(mpc):
    return (mpc * 1000 * u.kpc * arcsec_per_kpc).to(u.deg).value

# ----------------------------
# QUASAR
# ----------------------------
quasar = SkyCoord(ra=357.138875*u.deg,
                  dec=-30.9028417*u.deg)

# ----------------------------
# LOAD IMAGE
# ----------------------------
img_hdul = fits.open(IMG_FILE, memmap=True)
data = img_hdul[0].data.astype(np.float32)
wcs = WCS(img_hdul[0].header)

# ----------------------------
# LOAD WEIGHT → FOOTPRINT MASK
# ----------------------------
wt = fits.open(WT_FILE)[0].data.astype(np.float32)

threshold = 0.003
mask = (wt > threshold) & np.isfinite(wt)

data_plot = data.copy()
data_plot[~mask] = np.nan

# ----------------------------
# LOAD REGION (DECam footprint)
# ----------------------------
regions = Regions.read(REG_FILE, format='ds9')

# ----------------------------
# LOAD SOURCES + REMOVE QSO
# ----------------------------
cat = np.loadtxt(CAT_FILE)

# DEBUG (optional)
print("Catalog shape:", cat.shape)
print("First rows:\n", cat[:5])

sources_all = SkyCoord(ra=cat[:,0]*u.deg,
                       dec=cat[:,1]*u.deg)

# separation from quasar
sep = sources_all.separation(quasar)

mask_not_qso = sep > 1 * u.arcsec

sources = sources_all[mask_not_qso]

# ----------------------------
# ✅ BRIGHTNESS SELECTION (Y-band)
# ----------------------------
# Assume column 2 = Y-band magnitude
mag_y = cat[:,2][mask_not_qso]

# select brightest 20%
threshold_mag = np.percentile(mag_y, 20)

bright_mask = mag_y < threshold_mag

sources_bright = sources[bright_mask]
sources_faint  = sources[~bright_mask]

# convert to pixels
x_bright, y_bright = wcs.world_to_pixel(sources_bright)
x_faint,  y_faint  = wcs.world_to_pixel(sources_faint)

# quasar pixel
x_qso, y_qso = wcs.world_to_pixel(quasar)

# debug
print("Closest source distance (arcsec):", np.min(sep.arcsec))
print("Removed near QSO:", np.sum(~mask_not_qso))
print("Bright sources:", len(x_bright))

# ----------------------------
# RADII
# ----------------------------
r5 = mpc_to_deg(5)
r10 = mpc_to_deg(10)

pixel_scale = np.abs(wcs.pixel_scale_matrix[1,1])
r5_pix = r5 / pixel_scale
r10_pix = r10 / pixel_scale

# ----------------------------
# IMAGE SCALING
# ----------------------------
zscale = ZScaleInterval()
vmin, vmax = zscale.get_limits(data_plot)

# ----------------------------
# PLOT
# ----------------------------
fig = plt.figure(figsize=(6,6))
ax = plt.subplot(projection=wcs)

ax.set_facecolor('white')

# image
ax.imshow(data_plot,
          origin='lower',
          cmap='gray_r',
          vmin=vmin,
          vmax=vmax)

# ----------------------------
# FOOTPRINT
# ----------------------------
for reg in regions:
    pix_reg = reg.to_pixel(wcs)
    pix_reg.plot(ax=ax, color='black', lw=2)

# ----------------------------
# RINGS
# ----------------------------
circle_outer = plt.Circle((x_qso, y_qso),
                          r10_pix,
                          color='black',
                          fill=False,
                          linewidth=2)

circle_inner = plt.Circle((x_qso, y_qso),
                          r5_pix,
                          color='black',
                          fill=False,
                          linestyle=':',
                          linewidth=2)

circle_core = plt.Circle((x_qso, y_qso),
                         r5_pix*0.35,
                         color='gray',
                         alpha=0.3)

ax.add_patch(circle_core)
ax.add_patch(circle_outer)
ax.add_patch(circle_inner)

# ----------------------------
# SOURCES (UPDATED)
# ----------------------------
# faint population
ax.scatter(x_faint, y_faint,
           s=15,
           color='red',
           alpha=0.3,
           edgecolors='none',
           zorder=4)

# bright Y-band sources
ax.scatter(x_bright, y_bright,
           s=70,
           color='yellow',
           edgecolors='black',
           linewidth=0.6,
           zorder=6)

# quasar
ax.scatter(x_qso, y_qso,
           marker='*',
           s=120,
           color='black',
           zorder=7)

# ----------------------------
# AXES STYLE
# ----------------------------
ax.set_xlabel("RA")
ax.set_ylabel("DEC")

ax.coords[0].set_major_formatter('hh:mm')
ax.coords[1].set_major_formatter('dd:mm')

ax.minorticks_on()
ax.tick_params(which='both', direction='in', width=1.2)
ax.tick_params(which='major', length=4)

ax.grid(False)

plt.tight_layout()
plt.show()
```
