```python

"""
Compute signal-to-noise ratio (SNR) at given coordinates using photutils
with an aperture radius of 2 arcsec.
"""

from astropy.io import fits
from astropy.wcs import WCS
from astropy.coordinates import SkyCoord
import astropy.units as u
from photutils.aperture import CircularAperture, CircularAnnulus, aperture_photometry
from astropy.wcs.utils import proj_plane_pixel_scales
import numpy as np
import pandas as pd

# ---------- Input ----------
fits_path = "/Users/aishwarya/Documents/Lyman_alpha/N964/mosaic_N964_cut.fits"

target_coords = [
    ("23:48:33.33", "-30:54:10.23"),
    ("23:50:39.46", "-31:47:26.41"),
    ("23:49:25.39", "-31:46:34.95"),
    ("23:49:45.53", "-31:39:01.46"),
    ("23:48:22.23", "-31:37:53.92"),
    ("23:48:17.08", "-31:30:18.94"),
    ("23:46:51.65", "-31:19:29.93"),
    ("23:52:04.95", "-31:16:20.60"),
    ("23:44:51.34", "-31:14:01.78"),
    ("23:49:09.11", "-31:11:23.68"),
    ("23:48:38.18", "-31:10:25.49"),
    ("23:52:11.44", "-31:10:09.82"),
    ("23:50:09.67", "-31:00:28.43"),
    ("23:52:50.08", "-30:59:23.05"),
    ("23:46:52.33", "-30:57:07.02"),
    ("23:52:39.27", "-30:51:50.73"),
    ("23:51:42.32", "-30:50:46.97"),
    ("23:45:30.76", "-30:50:42.20"),
    ("23:45:38.56", "-30:41:06.23"),
    ("23:50:56.96", "-30:40:16.95"),
    ("23:46:20.25", "-30:40:03.94"),
    ("23:48:12.27", "-30:32:34.32"),
    ("23:44:45.07", "-30:31:07.12"),
    ("23:47:37.50", "-30:28:28.35"),
    ("23:44:31.64", "-30:27:35.79"),
    ("23:47:38.55", "-30:25:47.75"),
    ("23:44:36.91", "-30:25:04.43"),
    ("23:48:21.02", "-30:24:04.19"),
    ("23:47:21.19", "-30:22:53.54"),
    ("23:46:27.28", "-30:21:51.57"),
    ("23:49:27.05", "-30:21:25.98"),
    ("23:50:27.20", "-30:21:13.39"),
    ("23:45:02.91", "-30:12:32.87"),
    ("23:49:34.65", "-30:12:40.49"),
    ("23:50:23.24", "-30:10:45.41"),
    ("23:50:45.07", "-30:05:13.35"),
    ("23:49:43.48", "-30:03:01.26"),
    ("23:50:36.54", "-30:01:46.79"),
    ("23:48:43.58", "-29:53:13.94"),
]

# ---------- Load FITS and WCS ----------
hdu = fits.open(fits_path)[0]
data = np.nan_to_num(hdu.data, nan=0.0)
wcs = WCS(hdu.header)

# ---------- Determine pixel scale from WCS ----------
pixel_scales_deg = proj_plane_pixel_scales(wcs)  # degrees/pixel
pixel_scale = pixel_scales_deg[0] * 3600.0       # arcsec/pixel
print(f"Pixel scale = {pixel_scale:.3f} arcsec/pixel")

# ---------- Aperture parameters ----------
aperture_radius_pix = 2.0 / pixel_scale
annulus_r_in_pix = aperture_radius_pix * 2.0  # Inner radius of annulus
annulus_r_out_pix = aperture_radius_pix * 3.0  # Outer radius of annulus
print(f"Aperture radius = {aperture_radius_pix:.2f} pixels")

# ---------- Convert Ra and Dec to pixel ----------
sky_coords = SkyCoord(target_coords, unit=(u.hourangle, u.deg))
x, y = wcs.world_to_pixel(sky_coords)

# ---------- Run photometry ----------
"""
Perform aperture photometry and compute SNR for each target coordinate.
Handles NaN values in the data and ensures robust background estimation.
"""
results = []

for (ra, dec), xi, yi in zip(target_coords, x, y):
    if np.isnan(xi) or np.isnan(yi):
        results.append((ra, dec, np.nan, np.nan))
        continue

    # Define aperture and annulus
    aperture = CircularAperture((xi, yi), r=aperture_radius_pix)
    annulus = CircularAnnulus((xi, yi), r_in=annulus_r_in_pix, r_out=annulus_r_out_pix)

    # Aperture flux
    ap_phot = aperture_photometry(data, aperture)['aperture_sum'][0]

    # Background statistics
    ann_mask = annulus.to_mask(method='exact') # Get mask for annulus
    ann_data = ann_mask.multiply(data)
    ann_data_1d = ann_data[ann_mask.data > 0] # 1D array of annulus data

    if len(ann_data_1d) < 10:
        bkg_median = np.nan
        bkg_std = np.nan
    else:
        bkg_median = np.median(ann_data_1d)
        bkg_std = np.std(ann_data_1d)

    # Compute flux and SNR
    bkg_total = bkg_median * aperture.area
    flux = ap_phot - bkg_total
    noise = bkg_std * np.sqrt(aperture.area)
    snr = flux / noise if noise > 0 else np.nan

    results.append((ra, dec, flux, snr))

# ---------- Save results ----------
df = pd.DataFrame(results, columns=['RA', 'Dec', 'Flux', 'SNR'])

print("\n--- Aperture Photometry Results ---")
print(df.to_string(index=False, float_format="%.2f"))

outpath = "/Users/aishwarya/Documents/Lyman_alpha/snr_results_N964.csv"
df.to_csv(outpath, index=False)
print(f"\nResults saved to: {outpath}")
```
