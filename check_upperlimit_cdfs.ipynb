```python
import pandas as pd
import numpy as np

# ============================================================
# FILE
# ============================================================
z_file = "/Users/aishwarya/Desktop/new_cdfs/cat/LBG_Z_degraded.cat"

# ============================================================
# TARGET COORDINATES
# ============================================================
target_ra = [
    52.0415505, 51.8831737, 52.7681736, 52.8377927, 53.1412409,
    53.6023823, 52.6673702, 52.6587213, 51.9971687, 51.6118495,
    51.8135832
]

target_dec = [
    -28.4647062, -28.5403554, -28.4166605, -28.3696427, -28.401831,
    -28.3401939, -28.2050651, -28.2205090, -27.4568818, -27.7091367,
    -28.0709152
]

tolerance = 1.0 / 3600  # 1 arcsec
limiting_sigma = 5

# ============================================================
# LOAD FILE
# ============================================================
df = pd.read_csv(z_file, sep='\s+', header=None)

df.columns = [
    "ID", "X", "Y",
    "RA", "DEC",
    "MAG_APER", "MAGERR_APER",
    "MAG_AUTO", "MAGERR_AUTO",
    "FLAGS", "AREA",
    "MAG_APER_CAL", "MAGERR_APER_CAL"
]

# ============================================================
# SEARCH ALL TARGETS
# ============================================================
for ra, dec in zip(target_ra, target_dec):

    dist = np.sqrt((df["RA"] - ra)**2 +
                   (df["DEC"] - dec)**2)

    idx = np.argmin(dist)
    closest_dist = dist.iloc[idx]

    print("\n====================================")
    print(f"Target RA: {ra}, DEC: {dec}")

    if closest_dist < tolerance:
        match = df.iloc[idx]

        mag = match["MAG_APER_CAL"]
        magerr = match["MAGERR_APER_CAL"]

        sigma = 1.0857 / magerr

        print("Match found")
        print(f"Separation: {closest_dist*3600:.3f} arcsec")

        print("\n--- Magnitudes ---")
        print(f"Degraded MAG_APER_CAL   : {mag}")
        print(f"MAGERR_APER_CAL         : {magerr}")
        print(f"Detection sigma (SNR)   : {sigma:.2f}")

        # classification
        if sigma >= limiting_sigma:
            print("Status: Detected (>=5 sigma)")
        else:
            print("Status: Upper limit (<5 sigma)")

    else:
        print("No match within tolerance")
        print(f"Closest distance: {closest_dist*3600:.3f} arcsec")
```

```python
import pandas as pd
import numpy as np

# ============================================================
# FILE
# ============================================================
z_file = "/Users/aishwarya/Desktop/new_cdfs/cat/LBG_Z_degraded.cat"

# ============================================================
# TARGET COORDINATES
# ============================================================
target_ra = [
    52.0415505, 51.8831737, 52.7681736, 52.8377927, 53.1412409,
    53.6023823, 52.6673702, 52.6587213, 51.9971687, 51.6118495,
    51.8135832
]

target_dec = [
    -28.4647062, -28.5403554, -28.4166605, -28.3696427, -28.401831,
    -28.3401939, -28.2050651, -28.2205090, -27.4568818, -27.7091367,
    -28.0709152
]

tolerance = 1.0 / 3600  # 1 arcsec
limiting_mag = 26.60    # 5 sigma limiting magnitude

# ============================================================
# LOAD FILE
# ============================================================
df = pd.read_csv(z_file, sep='\s+', header=None)

df.columns = [
    "ID", "X", "Y",
    "RA", "DEC",
    "MAG_APER", "MAGERR_APER",
    "MAG_AUTO", "MAGERR_AUTO",
    "FLAGS", "AREA",
    "MAG_APER_CAL", "MAGERR_APER_CAL"
]

# ============================================================
# SEARCH ALL TARGETS
# ============================================================
for ra, dec in zip(target_ra, target_dec):

    dist = np.sqrt((df["RA"] - ra)**2 +
                   (df["DEC"] - dec)**2)

    idx = np.argmin(dist)
    closest_dist = dist.iloc[idx]

    print("\n====================================")
    print(f"Target RA: {ra}, DEC: {dec}")

    if closest_dist < tolerance:
        match = df.iloc[idx]

        mag = match["MAG_APER_CAL"]

        print("Match found")
        print(f"Separation: {closest_dist*3600:.3f} arcsec")

        # ----------------------------------------------------
        # CONDITION BASED ON LIMITING MAG
        # ----------------------------------------------------
        if mag > limiting_mag:

            print("\nUpper limit source (Below 5σ detection)")
            print(f"RA  : {match['RA']}")
            print(f"DEC : {match['DEC']}")
            print(f"MAG_APER_CAL : {mag}")
            print(f"MAGERR_APER_CAL : {match['MAGERR_APER_CAL']}")

        else:

            print("\nDetected source (Above 5σ)")
            print(f"RA  : {match['RA']}")
            print(f"DEC : {match['DEC']}")

    else:
        print("No match within tolerance")
        print(f"Closest distance: {closest_dist*3600:.3f} arcsec")
```

```python
import pandas as pd
import numpy as np

# ============================================================
# FILE PATHS
# ============================================================
main_file = "/Users/aishwarya/Desktop/CDFS_candidates_ZY_degraded.cat"
clean_file = "/Users/aishwarya/Desktop/source_cdfs_clean.cat"
output_file = "/Users/aishwarya/Desktop/CDFS_filtered.cat"

# ============================================================
# READ FILES CORRECTLY
# ============================================================

# Main file → comma separated
main_df = pd.read_csv(main_file)

# Clean file → whitespace separated
clean_df = pd.read_csv(clean_file, sep='\s+')

# ============================================================
# EXTRACT RA/DEC
# ============================================================

main_ra = main_df['RA']
main_dec = main_df['DEC']

clean_ra = clean_df['ALPHA_J2000']
clean_dec = clean_df['DELTA_J2000']

# ============================================================
# MATCHING (REMOVE these sources)
# ============================================================

tol = 1e-4   # ~0.36 arcsec

remove_mask = np.zeros(len(main_df), dtype=bool)

for ra_c, dec_c in zip(clean_ra, clean_dec):
    match = (np.abs(main_ra - ra_c) < tol) & (np.abs(main_dec - dec_c) < tol)
    remove_mask |= match

# ============================================================
# KEEP ONLY NON-MATCHED SOURCES
# ============================================================

filtered_df = main_df.loc[~remove_mask, ['RA', 'DEC']]

# Save in same format (comma-separated)
filtered_df.to_csv(output_file, sep=' ', index=False, header=False)

print(f"Removed {remove_mask.sum()} sources")
print(f"Saved {len(filtered_df)} sources to {output_file}")
```
